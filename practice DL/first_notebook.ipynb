```python
from keras.layers import *
```

```python
x = Input(shape=(3,))

```

```python
hidden1 = Dense(128,activation='relu')(x)
hidden2 = Dense(64,activation='relu')(hidden1)

output1 = Dense(1,activation='linear')(hidden2)
output2 = Dense(1,activation='sigmoid')(hidden2)
```

```python
model = Model(inputs = x ,outputs = [output1,output2])
```

```python
model.summary()
```

```python
from keras.models import Model

from keras.layers import *

deep_input = Input(shape=(392,))
wide_input = Input(shape=(392,))

dense_2 = Dense(30, activation='relu')(x)
dense_3 = Dense(30, activation='relu')(dense_2)
concatenate_1 = Concatenate()([dense_3, wide_input])


output = Dense(10, activation='softmax')(concatenate_1)

model = Model(inputs=x, outputs=output)

model.summary()

from keras.utils import plot_model
plot_model(model, show_shapes=True)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Concatenate

deep_input = Input(shape=(392,), name="deep_input")
wide_input = Input(shape=(392,), name="wide_input")

dense_2 = Dense(30, activation="relu", name="dense_2")(deep_input)
dense_3 = Dense(30, activation="relu", name="dense_3")(dense_2)
concat = Concatenate(name="concatenate_1")([wide_input, dense_3])
output = Dense(10, activation="softmax", name="output")(concat)

model = Model(inputs=[deep_input, wide_input], outputs=[output])
model.summary()

import tensorflow as tf
from tensorflow.keras.utils import plot_model
from IPython.display import Image

plot_model(
    model,
    to_file='model_architecture.png',
    show_shapes=True,        # Displays input/output shapes for each layer
    show_layer_names=True,   # Shows the names we assigned (e.g., 'dense_2')
    rankdir='TB',            # 'TB' for Top-to-Bottom, 'LR' for Left-to-Right
    dpi=96                   # Adjusts resolution/size of the image
)

Image('model_architecture.png')
```
